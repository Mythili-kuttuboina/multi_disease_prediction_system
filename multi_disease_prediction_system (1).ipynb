# Feature Engineering

```python
def BMI(x):
    if x < 18.5:
        return 'Underweight'
    elif x < 25:
        return 'Normal'
    elif x < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['BMI'].apply(BMI)
```

```python
def BP_Category(bp):
    if bp < 120 and bp< 80:
        return 0   # Normal
    elif bp < 140:
        return 1   # Pre-Hypertension
    else:
        return 2   # Hypertension

df['BP_Category'] = df['BloodPressure'].apply(BP_Category)
```

```python
def chol_risk(x):
    if x < 200:
        return 0   # Normal
    elif x < 240:
        return 1   # Borderline
    else:
        return 2   # High
df['Cholesterol_Risk'] = df['Cholesterol'].apply(chol_risk)
```
